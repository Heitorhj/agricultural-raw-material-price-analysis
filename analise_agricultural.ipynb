```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('agricultural_raw_material.csv')
df
```

```python
df.info()
```

```python
# Visualizando a quantidade de registros nulos

df.isnull().sum()
```

```python
# Tratando valores ausentes (substituindo eles pela mensagem abaixo)
df = df.dropna()
```

```python
# Conferindo se os valores nulos foram tratados com precisão
df.isna().sum()
```

```python
# Transformando dados
df = df.replace('%', '', regex=True)
df = df.replace(',', '', regex=True)
df = df.replace('-', '', regex=True)
df = df.replace('', np.nan) # => Troca todas as strings vazias por NaN
df = df.replace('MAY90', np.nan) # => Substitui esse texto por NaN
```

```python
# Convertendo os dados para tipo float
conversao_float = ["Coarse wool Price", "Coarse wool price % Change", "Copra Price", "Copra price % Change", "Cotton price % Change","Fine wool Price", "Fine wool price % Change", "Hard log price % Change", "Hard sawnwood price % Change", "Hide price % change", "Plywood price % Change", "Rubber price % Change", "Softlog price % Change", "Soft sawnwood price % Change", "Wood pulp price % Change"]
df[conversao_float] = df[conversao_float].astype('float')
df.dtypes
```

```python
# Traz as 5 primeiras linhas
df.head()
```

```python
# Colunas de data e hora (transformação)
df.Month = pd.to_datetime(df.Month.str.upper(), format='%b%y', yearfirst=False)
df = df.set_index('Month')
```

```python
df.head()
```

* Análise Exploratória

```python
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline

sns.set_style('darkgrid')

plt.rcParams['font.size'] = 14
plt.rcParams['figure.figsize'] = (9, 5)
plt.rcParams['figure.facecolor'] = 'white'
```

* Mapa de Calor

```python
# Escolhendo os materiais para analisar
Materiais_escolhidos = ['Coarse wool Price',  'Copra Price','Cotton Price', 'Fine wool Price',  'Hard log Price', 'Hard sawnwood Price',
 'Hide Price', 'Plywood Price', 'Rubber Price', 'Softlog Price', 'Soft sawnwood Price', 'Wood pulp Price']

# Aplicando o cálculo de correlação (estatística)
correlacao_materiais = df[Materiais_escolhidos].corr()

# configurando o tamanho do plot
fig = plt.figure(figsize= (12, 9))

# mascarando a parte superior do triângulo, escondendo assim a metade da matriz
mask = np.triu(np.ones_like(correlacao_materiais, dtype=bool))

sns.heatmap(
    correlacao_materiais,
    vmax= .8, # => Desenha com as cores na sua máxima intensidade
    mask=mask, # => deixa mascarada a metade da matriz
    square= True, # => Faz cada célula ficar quadrada
    annot=True) # => Escreve os números dentro das células
plt.show()
```

```python
# Plotando gráficos de linhas

ax = df[['Coarse wool Price', 'Coarse wool price % Change']].plot(
    figsize= (11, 9),
    subplots=True,
    linewidth = 2,
    grid=True,
    legend=True)

# Personalizando o primeiro gráfico
ax[0].set_title('Preço da Lã Grossa')
ax[0].set_ylabel('Preço (US$)')

# Personalizando o segundo gráfico
ax[1].set_title('Variação Percentual do Preço')
ax[1].set_ylabel('Variação (%)')
ax[1].set_xlabel('Ano')
plt.show()
```

```python
# Fazendo a Análise da Variação Normal do Preço de Cada Matéria-Prima
Va_materiaPrima = ['Copra price % Change','Softlog price % Change','Rubber price % Change','Cotton price % Change','Coarse wool price % Change','Fine wool price % Change','Hard log price % Change','Hard sawnwood price % Change','Hide price % change','Plywood price % Change','Soft sawnwood price % Change','Wood pulp price % Change']

for i in range(len(Va_materiaPrima)):
    plt.figure(figsize=(12,12))
    
    df[Va_materiaPrima[i]].hist(
        figsize=(11,9),
        edgecolor = 'black',
        linewidth = 1,)
    
    plt.title(f'Distribuição da Variação Percentual')
    plt.xlabel('Variação Percentual (%)')
    plt.ylabel('Frequência')
    plt.legend(Va_materiaPrima[i:], loc='upper center', bbox_to_anchor = (1.2,1))
```
